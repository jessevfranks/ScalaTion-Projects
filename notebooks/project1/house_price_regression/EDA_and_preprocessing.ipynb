```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

df = pd.read_csv('../../../data/project1/house_price_regression_dataset_raw.csv')
```

```python
print(df.head())
print(df.describe())
print(df.info())
```

Based on the summary above, we can see that there are not null values, and that no outliers need to be dealt with. Luckily, this make data preprocessing easy, as the dataset appears to be clean already and further processing isnt required.

```python
corr_matrix = df.corr()
sm.graphics.plot_corr(corr_matrix, xnames=corr_matrix.columns)
plt.show()
```

```python
# no NaN are present so we dont need to handle this
for col in df.columns:
    plt.figure(figsize=(6, 4))

    sm.qqplot(df[col], line='s')

    plt.title(f'QQ Plot: {col}')
    plt.show()
```
